```python
# 대상 파일 경로
file_path = "/content/drive/MyDrive/tft/libs/tft_model.py"

# 파일 읽고 수정하기
with open(file_path, 'r') as file:
    code = file.read()

# 'tf.variable_scope'를 'tf.name_scope'로 교체
updated_code = code.replace("tf.variable_scope", "tf.name_scope")

# 수정된 코드 저장하기
with open(file_path, 'w') as file:
    file.write(updated_code)

print(f"파일 {file_path}이 성공적으로 수정되었습니다.")
```
